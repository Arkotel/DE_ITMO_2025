```python
import pandas as pd

FILE_ID = "1gJrXyvqIVSZCEjqhGhvisyMyxI0zBald"  # ID файла на Google Drive
file_url = f"https://drive.google.com/uc?id={FILE_ID}"

raw_data = pd.read_csv(file_url, sep=";")  # читаем файл

print("\nТипы данных до обработки:")
print(raw_data.info())

# --- Создаем копию для обработки ---
proc_data = raw_data.copy(deep=True)
print(f"Размер исходных данных: {raw_data.shape}")
print(f"Размер созданной копии: {proc_data.shape}")

'''for column in proc_data.columns:
    missing_count = proc_data[column].isna().sum()
    unique_count = proc_data[column].nunique()
    
    if missing_count > 0 or unique_count in [2, 3]: #Визуально отображаем признаки, где есть пропущенные и/или 2-3 уникальных значения
        print(f"\n Колонка: {column}")
        print(f"   Тип: {proc_data[column].dtype}")
        print(f"   Уникальных значений: {unique_count}")
        print(f"   Значения: {list(proc_data[column].unique())}")
        print(f"   Пропусков: {missing_count}")
        value_counts = proc_data[column].value_counts(dropna=False)
        for value, count in value_counts.items():
            print(f"   '{value}': {count}")
'''

# --- Замена данных, изменение типов ---

proc_data['Speed_Exceeded'] = proc_data['Recorded_Speed'] > proc_data['Speed_Limit']
proc_data = proc_data.drop(['Speed_Limit', 'Recorded_Speed'], axis=1)
proc_data['Towed'] = proc_data['Towed'].map({'Yes': True, 'No': False}).astype(bool)
proc_data['Fine_Paid'] = proc_data['Fine_Paid'].map({'Yes': True, 'No': False}).astype(bool)
proc_data['Court_Appearance_Required'] = proc_data['Court_Appearance_Required'].map({'Yes': True, 'No': False}).astype(bool)
proc_data = proc_data.drop('Comments', axis=1) # Удаляем столбец Comments
proc_data = proc_data.drop('Alcohol_Level', axis=1) # Удаляем столбец Alcohol_Level
proc_data['Violation_Type'] = proc_data['Violation_Type'].astype('category')
proc_data['Fine_Amount'] = proc_data['Fine_Amount'].astype('int16')
proc_data['Location'] = proc_data['Location'].astype('category')
proc_data['DateTime'] = pd.to_datetime(proc_data['Date'] + ' ' + proc_data['Time'], format='%d.%m.%Y %H:%M') #Объединяем столбцы
proc_data = proc_data.drop(['Date', 'Time'], axis=1) # Удаляем столбцы Data и Time
proc_data['Vehicle_Type'] = proc_data['Vehicle_Type'].astype('category')
proc_data['Vehicle_Color'] = proc_data['Vehicle_Color'].astype('category')
proc_data['Vehicle_Model_Year'] = proc_data['Vehicle_Model_Year'].astype('int16')
proc_data['Registration_State'] = proc_data['Registration_State'].astype('category')
proc_data['Driver_Age'] = proc_data['Driver_Age'].astype('int8')
proc_data['Driver_Gender'] = proc_data['Driver_Gender'].astype('category')
proc_data['License_Type'] = proc_data['License_Type'].astype('category')
proc_data['Penalty_Points'] = proc_data['Penalty_Points'].astype('int8')
proc_data['Weather_Condition'] = proc_data['Weather_Condition'].astype('category')
proc_data['Road_Condition'] = proc_data['Road_Condition'].astype('category')
proc_data['Issuing_Agency'] = proc_data['Issuing_Agency'].astype('category')
proc_data['License_Validity'] = proc_data['License_Validity'].astype('category')
proc_data['Number_of_Passengers'] = proc_data['Number_of_Passengers'].astype('int8')

# Создаем понятные категории для Helmet_Worn
helmet_mapping = {'Yes': 'worn',           # надет
    'No': 'not_worn',        # не надет  
    None: 'not_required'     # не предусмотрен
}
proc_data['Helmet_Worn'] = proc_data['Helmet_Worn'].map(helmet_mapping).astype('category')
# Создаем понятные категории для Seatbelt_Worn
seatbelt_mapping = {
    'Yes': 'worn',           # надет
    'No': 'not_worn',        # не надет  
    None: 'not_required'     # не предусмотрен
}
proc_data['Seatbelt_Worn'] = proc_data['Seatbelt_Worn'].map(seatbelt_mapping).astype('category')

proc_data['Traffic_Light_Status'] = proc_data['Traffic_Light_Status'].astype('category')
proc_data['Breathalyzer_Result'] = proc_data['Breathalyzer_Result'].astype('category')
proc_data['Payment_Method'] = proc_data['Payment_Method'].astype('category')
proc_data['Previous_Violations'] = proc_data['Previous_Violations'].astype('int8')

print("\nТипы данных после обработки:")
print(proc_data.info())
```
